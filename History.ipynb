## Imports

```python
import os
from itertools import cycle

import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

## Learning curves

```python
root = '/data/lisatmp4/jegousim/iterative_inference/camvid/'  # Simon
# root = '' # Adriana

def extract_costs(line):
    tmp = line.split('cost train ')[1]
    cost_train, cost_val = tmp.split(', cost val ')[:2]
    cost_val = cost_val.split(' took')[0]
    return float(cost_train), float(cost_val)

def plot(folder_list):
    plt.figure(figsize=(10, 10))
    colors = cycle(['b', 'g', 'r', 'm', 'y', 'k', 'c'])
    for folder in folder_list:
        c = colors.next()
        with open(os.path.join(root, folder, 'output.log')) as f:
            lines = f.readlines()
            costs_list = [extract_costs(line) for line in lines]
            cost_train_list, cost_val_list = zip(*costs_list)
            plt.plot(cost_train_list, '-', color=c, label=folder)
            plt.plot(cost_val_list, '--', color=c)
    plt.legend()
```

```python
folder_list = [
'pool5',
]

plot(folder_list)
```
